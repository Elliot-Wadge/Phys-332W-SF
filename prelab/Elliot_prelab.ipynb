```
import numpy as np
import matplotlib.pyplot as plt
```

```
def fourier_plane(theta,wavelength,N,b,a,I = 1):

    k = np.pi/wavelength
    beta = (k*b)*np.sin(theta)
    alpha = (k*a)*np.sin(theta)
    return I/N**2 * (np.sin(beta)/beta)**2 * (np.sin(N*alpha)/np.sin(alpha))**2

def single(theta,wavelength, b):
    k = np.pi/wavelength
    beta = (k*b)*np.sin(theta)
   
    return (np.sin(beta)/beta)**2


b = 0.2e-3 #mm/line
a = 4*b
wavelength = 515e-9

Ns = np.arange(1,7,1)
theta = np.linspace(-3*wavelength/a,3*wavelength/a,10000)

fig, (ax1,ax2,ax3,ax4,ax5) = plt.subplots(5,1, figsize = (10,10))

axs = [ax1,ax2,ax3,ax4,ax5]


for ax,N in zip(axs,Ns):
    ax.plot(theta, fourier_plane(theta, wavelength, N, b, a), label = f"N = {N}")
    ax.legend()
    

plt.legend()
plt.show()

# plt.plot(theta, single(theta, wavelength, b))

```
